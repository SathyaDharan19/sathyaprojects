```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cl = pd.read_csv("wine-clustering.csv")
cl
```

```python
cl.info()
```

```python
cl.describe()
```

```python
x = cl["Alcohol"].replace(12.,12.00)
y = cl.Malic_Acid
print(x)
print(y)

z = list(zip(x,y))
print(z)
    
```

```python
a = []
for i in z:
    a.append(list(i))
    
print(a)    
    
```

```python
plt.scatter(x,y)
plt.xlabel('Alcohol amnt')
plt.ylabel('Malic acid amnt')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
qua = StandardScaler()
f = qua.fit(a)
print(f)
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(a)
y_kmeans = kmeans.predict(a)
```

```python
plt.scatter(x, y, c=y_kmeans, s=20)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='blue', s=100, alpha=0.9);
plt.show()
```

```python
cl["wine type"] = pd.Series(y_kmeans, index = cl.index)
```
